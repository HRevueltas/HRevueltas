```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
ds = sp.io.loadmat('ds00.mat')
```

```python
a = ds['trainingx'].T
b = ds['trainingy'].T
```

```python
fig, ax = plt.subplots()
ax.plot(a,b, "g.", label = 'dataset')
ax.set_title('Nonlinear Regression')
ax.legend()
ax.set_xlabel('$a$')
ax.set_ylabel('$b$')
fig.show()
```
